#### 6.- Realiza una función que realice una digestión con las siguientes enzimas de restricción en nuestra secuencia. La función tiene que devolver la siguiente información: (2 pto)
1. Enzimas que tienen sitios de corte en la secuencia y posición (o posiciones de corte)
2. Enzimas que no tienen sitios de corte (si las hay)
3. Resultado de la digestión: los fragmentos resultantes de nuestra secuencia
4. Indicar número y tamaño de cada fragmento

```python
enzR = {
    'HindIII' : 'A*AGCTT',
    'BamHI' : 'G*GATCC',
    'AluI' : 'AG*CT',
    'Sau3AI' : '*GATC',
    'EcoRI' : 'G*AATTC'
}
```

```python
# Función digestión ADN con enzs restriccion
def digestion_restriccion(secuencia, enzimas):
    # Limpiar: saltos de línea, espacios, en mayusc
    seq = secuencia.replace("\n", "").replace(" ", "").upper()

    cortes = {}        # Enzs que cortan
    no_cortan = []     # Enzs que no 
    fragmentos = []    # Fragmentos

    # Buscar sitios de corte de cada enz
    for nombre, patron in enzimas.items():
        # Determinar posición corte con *
        corte_idx = patron.index('*')
        sitio = patron.replace('*', '')  # sitio  reconocimiento sin asterisco
        
        posiciones = []
        start = 0
        while True:
            pos = seq.find(sitio, start)
            if pos == -1:
                break
            posiciones.append(pos + corte_idx)  # posición real de los cortes
            start = pos + 1
        if posiciones:
            cortes[nombre] = posiciones
        else:
            no_cortan.append(nombre)

    # fragmentos ADN simulando digestión
    posiciones_corte = sorted([p for lista in cortes.values() for p in lista])
    inicio = 0
    for pos in posiciones_corte:
        fragmentos.append(seq[inicio:pos])
        inicio = pos
    fragmentos.append(seq[inicio:])  # Último fragmento

     # Mostrar
    print("Enzimas que cortan y posiciones:")
    for enzima, pos in cortes.items():
        print(f" - {enzima}: {pos}")
    
    if no_cortan:
        print("\nEnzimas que no cortan:", ", ".join(no_cortan))
    else:
        print("\nTodas las enzimas tienen al menos un sitio de corte.")

    print(f"\nNúmero total de fragmentos: {len(fragmentos)}")
    print("Tamaños de los primeros 10 fragmentos:", [len(f) for f in fragmentos[:10]])

    # Devolver
    return cortes, no_cortan, fragmentos
```

```python
# Llamar función secuencia de ADN
cortes, no_cortan, fragmentos = digestion_restriccion(secuencia, enzR)
```

#### 7.- Realiza una función que mediante la introducción de una secuencia de ADN devuelva lo siguiente: (3pto)
1. Todos los ORFs posibles que puedan surgir de la secuencia.
2. Selecciona los 6 ORFs más largos e indica si corresponden a la strand positiva o negativa.
3. Devuelve como mejor predicción, el más largo de todos.
4. Calcula el porcentaje de aminoácidos aromáticos de esa proteína elegida y su longitud relativa a la secuencia original de ADN.

```python
# Tabla de codones ADN -> aminoácidos
codon_table = {
    'ATA':'I', 'ATC':'I', 'ATT':'I', 'ATG':'M',
    'ACA':'T', 'ACC':'T', 'ACG':'T', 'ACT':'T',
    'AAC':'N', 'AAT':'N', 'AAA':'K', 'AAG':'K',
    'AGC':'S', 'AGT':'S', 'AGA':'R', 'AGG':'R',
    'CTA':'L', 'CTC':'L', 'CTG':'L', 'CTT':'L',
    'CCA':'P', 'CCC':'P', 'CCG':'P', 'CCT':'P',
    'CAC':'H', 'CAT':'H', 'CAA':'Q', 'CAG':'Q',
    'CGA':'R', 'CGC':'R', 'CGG':'R', 'CGT':'R',
    'GTA':'V', 'GTC':'V', 'GTG':'V', 'GTT':'V',
    'GCA':'A', 'GCC':'A', 'GCG':'A', 'GCT':'A',
    'GAC':'D', 'GAT':'D', 'GAA':'E', 'GAG':'E',
    'GGA':'G', 'GGC':'G', 'GGG':'G', 'GGT':'G',
    'TCA':'S', 'TCC':'S', 'TCG':'S', 'TCT':'S',
    'TTC':'F', 'TTT':'F', 'TTA':'L', 'TTG':'L',
    'TAC':'Y', 'TAT':'Y', 'TAA':'_', 'TAG':'_',
    'TGC':'C', 'TGT':'C', 'TGA':'_', 'TGG':'W'
}

def analizar_ORFs(secuencia):
    # Función hebra complementaria
    def complementaria(seq):
        complement = {'A':'T','T':'A','C':'G','G':'C'}
        return ''.join([complement[base] for base in seq])

    # Función traduccion ADN a proteína
    def traducir(seq):
        prot = ''
        for i in range(0, len(seq)-2, 3):
            codon = seq[i:i+3]
            prot += codon_table.get(codon, 'X')  # 'X' si codon desconocido
        return prot

    # Generar las 6 traducciones posibles (3 de +, 3 de -)
    strands = {'+': secuencia, '-': complementaria(secuencia)[::-1]}
    orfs = []

    for strand_name, seq in strands.items():
        for frame in range(3):
            prot = traducir(seq[frame:])
            # Dividir en ORFs: secuencias entre M y stop (_)
            inicio = 0
            while inicio < len(prot):
                if prot[inicio] == 'M':
                    fin = inicio
                    while fin < len(prot) and prot[fin] != '_':
                        fin += 1
                    orfs.append({'protein': prot[inicio:fin], 'strand': strand_name})
                    inicio = fin
                else:
                    inicio += 1

    # Ordenar ORFs por longitud descendente
    orfs.sort(key=lambda x: len(x['protein']), reverse=True)

    # Seleccionar los 6 más largos
    top6 = orfs[:6]

    # Mejor predicción: el ORF más largo
    mejor = top6[0] if top6 else None

    # Calcular % de aa aromáticos (F, Y, W) y longitud relativa
    if mejor:
        arom = sum(mejor['protein'].count(a) for a in ['F','Y','W'])
        porcentaje_aromaticos = (arom / len(mejor['protein'])) * 100 if len(mejor['protein'])>0 else 0
        longitud_relativa = (len(mejor['protein'])*3 / len(secuencia)) * 100  # proteína -> bases ADN
        mejor['porcentaje_aromaticos'] = porcentaje_aromaticos
        mejor['longitud_relativa'] = longitud_relativa

    return orfs, top6, mejor
```

```python
# Llamar función secuencia limpia
orfs, top6, mejor_proteina = analizar_ORFs(secuencia)

# Mostrar 6 ORFs más largos
print("6 ORFs más largos:")
for i, o in enumerate(top6, 1):
    print(f"{i}. Longitud: {len(o['protein'])} aa, Strand: {o['strand']}")

# Mejor predicción
if mejor_proteina:
    print("\nMejor predicción (ORF más largo):")
    print("Longitud:", len(mejor_proteina['protein']), "aa")
    print("Strand:", mejor_proteina['strand'])
    print("Porcentaje de aminoácidos aromáticos:", round(mejor_proteina['porcentaje_aromaticos'],2), "%")
    print("Longitud relativa respecto al ADN:", round(mejor_proteina['longitud_relativa'],2), "%")
```
